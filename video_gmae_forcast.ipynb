## 3.2 Platform Sales Analysis Over Time

Now let's analyze how sales vary across platforms and years:

<div class="alert alert-info">
<b> Student's comment</b>

create a new df grouping platform and year_of_release by total sales.
</div>


````python
# Calculate total sales by platform and year

sales_by_platform = df_filtered.groupby(['platform', 'year_of_release'])['total_sales'].sum().reset_index()
print(sales_by_platform)


````

````python
# Create a heatmap of platform sales over time

platform_heatmap = sales_by_platform.pivot(index='year_of_release', columns='platform', values='total_sales')
plt.figure(figsize=(15, 10))
sn.heatmap(platform_heatmap, annot=False, cmap='YlOrRd',  
            cbar_kws={'label': 'Total Sales (millions)'})
plt.title('Platform Sales Over Time (2000 - 2016)')
plt.xlabel('platform')
plt.ylabel('Year Of Release')
plt.xticks(rotation=45)
plt.show()
````

<div class="alert alert-info">
<b> Student's comment</b>

With the heat map we can see how sales decline as the platforms phase out with the next generation coming in. The one that stays reletively constant is the PC. Even though technology advances, consumers can upgrade thier PC to evolve with the new technology requirements.
</div>


````python
# Identify platforms with declining sales
platform_trends = []

for platform in sales_by_platform['platform'].unique():
    platform_data = sales_by_platform[sales_by_platform['platform'] == platform].sort_values('year_of_release')
    
    if len(platform_data) >= 3:  
        years = platform_data['year_of_release'].values
        sales = platform_data['total_sales'].values
        
       
        n = len(years)
        slope = (n * np.sum(years * sales) - np.sum(years) * np.sum(sales)) / (n * np.sum(years**2) - np.sum(years)**2)
        
        platform_trends.append({
            'platform': platform,
            'slope': slope,
            'total_sales': platform_data['total_sales'].sum(),
            'years_active': len(platform_data),
            'first_year': platform_data['year_of_release'].min(),
            'last_year': platform_data['year_of_release'].max()
        })
trends_df = pd.DataFrame(platform_trends).sort_values('slope')
print("Platform Sales Trends (2006-2016):")
print("Negative slope = Declining, Positive slope = Growing")
print(trends_df)

declining_platforms = trends_df[trends_df['slope'] < -5]  # Declining by more than 5 million per year
print("\nPlatforms with Significant Declining Sales:")
print(declining_platforms[['platform', 'slope', 'total_sales', 'first_year', 'last_year']])
````

# Step 5 : Hypothesis Tests

—Average user ratings of the Xbox One and PC platforms are the same.

—Average user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

—How you formulated the null and alternative hypotheses

—What criteria you used to test the hypotheses~~,~~ and why


````python
xone_rating = df_relevant.loc[df_relevant['platform'] == 'XOne', 'user_score']
pc_rating = df_relevant.loc[df_relevant['platform'] == 'PC', 'user_score']

alpha = 0.05
results = st.ttest_ind(
    xone_rating, 
    pc_rating,
    equal_var=False,
    nan_policy='omit')

print('p-value:', results.pvalue)

if (results.pvalue) < alpha:
    print("The user rating for Xbox One and PC are different")
else:
    print("We don't have enough evedince to conclude the ratings differ")
````


<div class="alert alert-info">
<b> Student's comment</b>

Null Hypothesis- The average User ratings of Xbox one and PC games are the same. Alternative hypothesis- The average user scores of Xbox One and PC are Different.
Used the standard alpha value to make it difficult to reject the null hypothesis. Chose a 2 tailed test to compare the 2 different platforms with thier user scores.
from the P value we can conclude the user scores are similar for the Xbox One and PC.
</div>


````python
action_rating = df_relevant.loc[df_relevant['genre'] == 'Action', 'user_score']
sports_rating = df_relevant.loc[df_relevant['genre'] == 'Sports', 'user_score']

alpha = 0.05
results = st.ttest_ind(
    action_rating, 
    sports_rating,
    equal_var=False,
    nan_policy='omit')

print('p-value:', results.pvalue)

if (results.pvalue) < alpha:
    print("The user rating for Actions and Sports are different")
else:
    print("We don't have enough evedince to conclude the ratings differ")
````
